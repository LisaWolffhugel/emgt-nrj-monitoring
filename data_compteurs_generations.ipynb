```python
from pandas import pandas
import datetime
import random
```

```python
type_compteur = pandas.read_csv("./data/V1/type_compteur.csv", delimiter=";")
```

```python
type_compteur.head(10)
```

```python
compteur_list = type_compteur["ID_compteur"].to_list()
```

```python
today = datetime.datetime.now()
```

```python
result_list = []
```

```python
for index, compteur_name in enumerate(compteur_list):
    # only data during 2 weeks (every day) for all meters of level 2 and 3 are generated not the level 1
    if index > 3: 
        for i in range(15): 
            result_list.append((compteur_name, (today+datetime.timedelta(days=-i)).strftime("%Y-%m-%d %H:%M:%S.000"), round(random.uniform(10,30), 2)))
    
```

```python
result_df = pandas.DataFrame(result_list,columns=["tagName", "ts", "tagValue"])
```

```python
result_df.head()
```

```python
result_df.to_csv("./data/compteurs.csv",sep=";",index=False)
```
